### 建立模型

```python
model = Model("LiNingExample")
```

### 变量及参数定义

```python
M = 10000   #极大值
diffabs = {} #绝对值
bl = {} #bool
X = {} #决策变量
Xmax = {} # X 的最大值
dllb = {}
diff = {}
```

### 决策变量X，每个尺码的件数

```python
dfx = data_result.reset_index('org_id')
for idx_s in dfx.index.unique():
    class_2, org_id = idx_s
#     print(idx_s)
    X[class_2, org_id] = model.addVar(vtype="I", name="X")
```

```python
print(len(X))
```

### 设置目标

```python
dfx
```

### 辅助变量 max 尺码件数X中的最大值Xmax

```python
dfmax = data_result.reset_index(['size','org_id'])
# dfmax
```

```python
for idx_s in dfx.index.unique():
    class_2, size = idx_s
    print(idx_s)
    idx_co = (class_2)
    print (idx_co)
   
    Xmax[idx_co] = model.addVar(vtype ='I', name = "Xmax")
    bl[idx_co] = model.addVar(vtype="B", name="bl")

    model.addCons(Xmax[idx_co] - X[idx_s] >= 0)
    model.addCons(X[idx_s] - Xmax[idx_co] + M * (1 - bl[idx_co])>= 0 )
    model.addCons(X[idx_s] - Xmax[idx_co] + M * bl[idx_co] <= 1 )
```

### 辅助等式  绝对值计算 及约束确定

```python
# data_result
```

```python
# for org_id in dfmax.index.unique():
#     print(org_id)
for idx in data_result.index.unique():
#     print(idx)
    myratio = data_result.loc[idx,'ratio']
    dllb[idx] = model.addVar(vtype = 'B', name = 'dllb')
    diff[idx] = model.addVar(vtype = 'C', name = 'diff', obj = data_result.loc[idx, 'qty'])
    diffabs[idx] = model.addVar(vtype = 'C', name = 'diffabs')

    class_2, org_id, size = idx
    
#     print(idx_x)
#     print(class_2, org_id)
    model.addCons(diff[idx] == X[class_2, size] - myratio * Xmax[class_2])

    model.addCons(diffabs[idx] >= diff[idx])
    model.addCons(diffabs[idx] >= -1 * diff[idx])
    model.addCons(diffabs[idx] <= diff[idx] + M * (1 - dllb[idx]))
    model.addCons(diffabs[idx] <= -1 * diff[idx] + M * dllb[idx])

#     model.addCons(diff[idx] + M * (1 - dllb[idx]) >= 0)
#     model.addCons(diff[idx] - M * dllb[idx] <= 1)


#     model.addCons(diffabs[idx] >= diff[idx])
#     model.addCons(diffabs[idx] >= -diff[idx])
#     model.addCons(diffabs[idx] <= diff[idx] + M * (1-dllb[idx]))
#     model.addCons(diffabs[idx] <= M * dllb[idx])
#绝对值
    
```
