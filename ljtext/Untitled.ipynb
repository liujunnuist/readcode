```python
proud_sql = '''select edw2.dim_product.product_id as prod_id,
	   edw2.dim_product_sku.color_id,
	   edw2.dim_product_sku.size_id,
	   edw2.dim_product_sku.order_num as size_order,
	   edw2.dim_product.year_id as year,
	   edw2.dim_product.quarter_id as season_id,
	   edw2.dim_product.big_class as class_0,
	   edw2.dim_product.mid_class as class_1,
	   edw2.dim_product.tiny_class as class_2,
	   edw2.dim_product.band as band_id,
	   edw2.dim_product.tag_price as price_tag
from edw2.dim_product
inner join edw2.dim_product_sku
on edw2.dim_product.product_id = edw2.dim_product_sku.product_id 
where edw2.dim_product.tiny_class='短袖T恤' and edw2.dim_product.year_id='2018' '''
proud = pd.read_sql_query(proud_sql, engine)
proud
```
