```python
def cal_clothes_model_1(data_result,minput):
    data_result = clo_data_test.set_index(['class_2','org_id','size','sex'])
#建立模型
    model = Model("LiNingExample")

    #  变量及参数定义
    M = 100000   #极大值
    diffabs = {} #绝对值
    bl = {} #bool
    X = {} #决策变量
    Xmax = {} # X 的最大值
    dllb = {}
    diff = {}

    #     辅助等式  绝对值变量
    for idx in data_result.index.unique(): #class org size 
#         print(idx)
        myratio = data_result.loc[idx,'ratio']
        # print('myration:', myratio)
        dllb[idx] = model.addVar(vtype = 'B', name = 'dllb')
        diff[idx] = model.addVar(lb = None,  vtype = 'C', name = 'diff')
        diffabs[idx] = model.addVar(vtype = 'C', name = 'diffabs')

    dfx = data_result.reset_index('org_id')
    # 决策变量X，每个尺码的件数
    for idx_s in dfx.index.unique(): #class size sex
        X[idx_s] = model.addVar(vtype="I", name="X")
    #     print(idx_s,'**********',X[idx_s])

    # 设置
    df_s = data_result.reset_index(['size'])  #class_2 org_id sex

    for class_2, org_id,sex in df_s.index.unique():
        #获取 class org ratio的ratio的最大值的尺码
        idx = (class_2, org_id, sex)
        org_xmax = df_s.loc[idx].sort_values(by = 'ratio',axis = 0,ascending = False).head(1).loc[idx,'size']
    #     print(org_xmax)
    #     print(org_xmax)
#         print(idx,'最大尺码-------------', org_xmax)
        po =df_s.loc[[(class_2, org_id, sex)]].reset_index().set_index(['class_2','org_id','size','sex']) #?多了一个index
        for idx in po.index.unique(): #class org size
#             print(idx)
            myratio = po.loc[idx,'ratio']
            class_2, org_id, size,sex = idx
    #             print('-------------org',org_id)
    #             print('-------------class size',class_2,size)
    #             print('*'*10, idx)
    #             print('diff:',diff[idx])
    #             print('X[class_2, size]:',X[class_2, size])
    #             print('X[class_2, org_max]:', X[class_2, org_max])
    #         print(class_2, org_xmax,sex)
            model.addCons(diff[idx] == X[class_2, size, sex] - myratio * X[class_2, org_xmax,sex])
            model.addCons(diffabs[idx] >= diff[idx])
            model.addCons(diffabs[idx] >= -1 * diff[idx])
            model.addCons(diffabs[idx] <= diff[idx] + M * (1 - dllb[idx]))
            model.addCons(diffabs[idx] <= -1 * diff[idx] + M * dllb[idx])  



    #     df_1 = data_result.reset_index(['org_id','size'])
    #     for idx in df_1.index.unique():
    #         class_2 = idx
    #         model.addCons(quicksum(bl[class_2, size_1] for size_1 in df_1.loc[idx, ['size']].drop_duplicates().values.ravel()) >= 1)

    #设置目标
    for idx in data_result.index.unique():
    #     print(idx)
        model.setObjective(data_result.loc[idx, 'nor']*diffabs[idx], clear = False)

    # 约束： 每个尺码的件数之和 等于 最小装箱数
    df = data_result.reset_index('size')
    for idx_s in df.index.unique(): #class_2	org_id	sex
    #     print(idx_s)
        class_2, org_id, sex = idx_s
    #     print('--------------------',class_2,org_id,sex)
    #     for size_1 in df.loc[idx_s, ['size']].values.ravel():
    #         print(size_1)
        model.addCons(quicksum(X[class_2,size_1,sex] for size_1 in df.loc[idx_s, ['size']].values.ravel()) == minput)


    model.setMinimize()
    model.optimize()
    status = model.getStatus()

    # 获取输出值
    OutX = {}
    if model.getStatus() =='optimal':
    #     print("out")
    #     print(X)
        for i in X:
            OutX[i] = round(model.getVal(X[i]))
    #     for i in X:
    #         print(i)
    # #         for idx11 in Xmax:
    # #             print('Xmax--',idx, '= ', model.getVal(Xmax[idx]),"======")
        liningResult = pd.DataFrame(OutX, index =['尺码件数']).T.rename_axis(['小类','尺码','']).reset_index()
        print(liningResult)

    else:
        print("status",'*'*30, model.getStatus())
        print("obj:", model.getObjVal())
        print("gap:", model.getGap())
    return OutX
```

```python
def cal_clothes_model_2(data_result):
    
    data_result = data_result.reset_index().set_index(['class_2','org_id','size'])
    
#     minput = 18
#     minput = 12
    
    #建立模型
    model = Model("LiNingExample")

    #  变量及参数定义
    M = 100000   #极大值
    diffabs = {} #绝对值
    bl = {} #bool
    X = {} #决策变量
    Xmax = {} # X 的最大值
    dllb = {}
    diff = {}
    
    dfx = data_result.reset_index('org_id')
    
    #决策变量X，每个尺码的比例
    for idx_s in dfx.index.unique():
        X[idx_s] = model.addVar(vtype="C", name="X")
        
    #辅助变量 max 尺码件数X中的最大值Xmax
    dfmax = data_result.reset_index(['size','org_id'])
    
    for idx_s in dfx.index.unique():
        class_2, size = idx_s
        if class_2 not in Xmax:
            Xmax[class_2] = model.addVar(vtype ='C', name = "Xmax")
        bl[idx_s] = model.addVar(vtype="B", name="bl")

        model.addCons(Xmax[class_2] - X[idx_s] >= 0)
        model.addCons(X[idx_s] - Xmax[class_2] + M * (1 - bl[idx_s])>= 0 )
        model.addCons(X[idx_s] - Xmax[class_2] - M * bl[idx_s] <= -1 )
    
    #辅助等式  绝对值计算 及约束确定
    for idx in data_result.index.unique():
        myratio = data_result.loc[idx,'ratio']
        # print('myration:', myratio)
#         if myratio!=0:
#             print(idx,'==========',myratio)
        dllb[idx] = model.addVar(vtype = 'B', name = 'dllb')
        diff[idx] = model.addVar(lb = None,  vtype = 'C', name = 'diff')
        diffabs[idx] = model.addVar(vtype = 'C', name = 'diffabs')
#       diffabs[idx] = model.addVar(vtype = 'C', name = 'diffabs', obj = data_result.loc[idx, 'qty'])
    
    #设置目标

    for idx in data_result.index.unique():
        model.setObjective(data_result.loc[idx, 'qty']*diffabs[idx], clear = False)
#     for idx in data_result.index.unique():
#         model.setObjective(data_result.loc[idx, 'qty']*diffabs[idx], clear = False)
# #         model.setObjective(diffabs[idx], clear = False)

    #绝对值
    for idx in data_result.index.unique():
        myratio = data_result.loc[idx,'ratio']
        class_2, org_id, size = idx
        
#         if myratio!=0:
#             print('==========',myratio)
        model.addCons(diff[idx] == X[class_2, size] - myratio * Xmax[class_2])

        model.addCons(diffabs[idx] >= diff[idx])
        model.addCons(diffabs[idx] >= -1 * diff[idx])
        model.addCons(diffabs[idx] <= diff[idx] + M * (1 - dllb[idx]))
        model.addCons(diffabs[idx] <= -1 * diff[idx] + M * dllb[idx])
        
        
#         class_2, org_id, size = idx
#         model.addCons(diff[idx] == X[class_2, size] - myratio * Xmax[class_2])
    
    df = data_result.reset_index('size')
    df_1 = data_result.reset_index(['org_id','size'])
    for idx in df_1.index.unique():
        class_2 = idx
        model.addCons(quicksum(bl[class_2, size_1] for size_1 in df_1.loc[idx, ['size']].drop_duplicates().values.ravel()) >= 1)
    
#     约束： 每个尺码的件数之和 等于 最小装箱数
#     for idx_s in df.index.unique():
#         class_2, org_id = idx_s
#         model.addCons(quicksum(X[class_2,size_1] for size_1 in df.loc[idx_s, ['size']].values.ravel()) == minput)

    
    model.setMinimize()
    model.optimize()
    status = model.getStatus()
    
    # 输出到出X为excle到本地
    OutX = {}
    if model.getStatus() =='optimal':
        for i in X:
            OutX[i] = round(model.getVal(X[i]))
        for idx11 in Xmax:
            print('Xmax--',idx, '= ', model.getVal(Xmax[idx]),"======")
    else:
        print("status",'*'*30, model.getStatus())
        print("obj:", model.getObjVal())
        print("gap:", model.getGap())
#         break
    return OutX
        
#     print(OutX)
#     print(liningResult)
    
#     return OutX
```
